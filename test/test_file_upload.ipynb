```python
from bluebook import token_manager, file_manager
from pathlib import Path

example_file = Path("./media/example-zero-trust.pdf")
example_file.exists()
```

```python
config = token_manager.load_config()
fm = file_manager.FileManager(config["API_TOKEN"])
```

```python
fm.upload_local_file(filepath=example_file)
```

```python
ufm = file_manager.UploadedFilesManager(token=config["API_TOKEN"])
ufm.list_files(names_only=True)
```

```python
uploaded = ufm.get_file(name="example-zero-trust")
uploaded.sha256_hash
```

```python
fm.local_media_manager.base64_sha256(filepath=fm.local_media_manager.get_local_filepath("example-zero-trust.pdf"))
```
