```python
from bluebook import token_manager, file_manager
from pathlib import Path

example_file = Path("./media/example-zero-trust.pdf")
example_file.exists()
```

```python
fm = file_manager.FileManager(token=token_manager.load_config()['API_TOKEN'])
```

```python
fm.upload_file(filepath=example_file, name="test-file")
```

```python
files = fm.list_files()

example_uploaded_files = files[0]
```

```python
list(files)
```

```python
print(example_uploaded_files.state)
print(example_uploaded_files.name)
print(example_uploaded_files.display_name)
```

```python
got_file = fm.get_file("files/test-file")
print(got_file.state)
print(got_file.name)
```

```python
client = fm.client
response = client.models.generate_content(
    model="gemini-2.0-flash",
    contents=["generate a multiple choice question from following pdf", got_file],
)
print(response.text)
```
