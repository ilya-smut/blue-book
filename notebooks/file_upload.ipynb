```python
from bluebook.file_manager import FileManager
from bluebook.configuration import Configuration

example_file_path = Configuration.LocalAppPath.BLUEBOOK_DIR.parent / "sample_files" / "file1.txt"
fm = FileManager()

print(example_file_path)
print(fm.ls_remote())
```

```python
fm.remove_from_remote('file1.txt')
print(fm.ls_remote())
```

```python
print(fm.ls_cache_dir(str_names=True))
fm.remove_from_cache("file1.txt")
print(fm.ls_cache_dir(str_names=True))
```

```python
fm.copy2cache(example_file_path)
fm.ls_cache_dir(str_names=True)
```

```python
fm.upload_from_cache(name="file1.txt")
remote_files = fm.ls_remote()
print(remote_files)
```

```python
remote_file = fm.get_from_remote(list(remote_files.keys())[0])
print(remote_file.display_name)
```

```python
attached_files = []
attached_files.append(remote_file)

from bluebook import generator, token_manager
token = token_manager.load_config()["API_TOKEN"]

response = generator.ask_gemini(exam_name="Driving Test UK", question_num=1, token=token, attached_files=attached_files)
```

```python
from bluebook import data_models

questions = data_models.serialize_questions(response)
print(questions)
```
