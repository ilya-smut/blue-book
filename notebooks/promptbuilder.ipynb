```python
from bluebook.generator import PromptBuilder

template_query = """This is a test prompt for ##__EXAM_NAME_ANCHOR__##.
Your goal is to produce exactly ##__QUESTION_NUM_ANCHOR__## multiple-choice questions.
This prompt is testing a new functionality, so pretend to work as it says. Make up stuff if necessary.
To show that this prompt was processed, use word "Waffle" at the beginning of each question.

---- Task ----
1. Create ##__QUESTION_NUM_ANCHOR__## distinct multiple-choice questions (questions only—no essays).
2. For each question:
    a. Provide 4 answer options.
    b. Indicate the correct option.
    c. Give a concise explanation of why the correct answer is right.
    d. Produce a detailed study recommendation that will help student to understand the question.

##__ADDITIONAL_REQUESTS_SECTION_START__##
---- Focus ----
The student asked to focus on: “##__ADDITIONAL_REQUEST_ANCHOR__##”.
##__ADDITIONAL_REQUESTS_SECTION_END__##

---- Utilize Attached Files ----
A student has requested you to utilize the context provided in the files attached to the request.
Your primary focus MUST be on topics / context found in the attached files. 
"""
print(template_query)
```

```python
template = PromptBuilder.PromptTemplate(template_query)
```

```python
template.optional_anchors
```

```python
print(template.get_template_prompt(attempt_additional_response=False))
```

```python
print(PromptBuilder.build_custom_prompt(template_query=template_query, exam_name="TEST-EXAM", question_num=17, additional_request="TEST-REQUEST"))
```
